### Dimentionality Reduction u/ Visualisasi
ada 128 feature yang digunakan u/ klasifikasi (128 titik landmark wajah). Tetapi tidak mungkin di-visualisasikan dengan 128 dimensi. Jadi kita reduksi menjadi 3 dimensi dengan PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
V = pca.fit_transform(df.drop('128', axis=1))

pca_df = pd.DataFrame(V, columns=['pc_1', 'pc_2', 'pc_3'])
pca_df['name'] = df['128']
pca_df
```

```python
groups = pca_df.groupby('name')
groups
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
import numpy as np
%matplotlib inline  


fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111, projection='3d') 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_zlabel('Principal Component 3', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)
targets = df['128']
colors = cm.rainbow(np.linspace(0, 1, len(targets)))

for target, color in zip(targets,colors):
# for target in targets:
    indicesToKeep = pca_df['name'] == target
    ax.scatter(pca_df.loc[indicesToKeep, 'pc_1']
               , pca_df.loc[indicesToKeep, 'pc_2']
               , pca_df.loc[indicesToKeep, 'pc_3']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
import seaborn as sns

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(15, 12))
sns.despine(f, left=True, bottom=True)

sns.scatterplot(x="pc_1", y="pc_2",
                hue="name",
                palette="deep",
                sizes=(1, 8), linewidth=0,
                data=pca_df, ax=ax)
```
